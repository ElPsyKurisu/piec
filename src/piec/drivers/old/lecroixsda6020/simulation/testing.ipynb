```python
import numpy as np

def capacitor_response(v, t, C):
    # Calculate the time step
    dt = np.diff(t)
    
    # Initialize the current array
    i = np.zeros_like(v)
    
    # Calculate the current using the capacitor equation I(t) = C * dV(t)/dt
    i[1:] = C * np.diff(v) / dt
    
    # Integrate the current to get the voltage response
    v_response = np.cumsum(i) * dt[0] / C
    
    return v_response

# Example usage
v = np.array([0, 1, 2, 3, 4, 5])  # Example voltage array
t = np.array([0, 1, 2, 3, 4, 5])  # Example time array
C = 1000.0  # Capacitance value

v_response = capacitor_response(v, t, C)
print(v_response)
```
